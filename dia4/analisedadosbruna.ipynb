#agrupando o banco a partir de raÃ§a e etnia e plotando grafico com as notas de matematica e a media

#plotando um grafico de barras para demonstrar a media em matematica de cada raÃ§a/etnia, demonstrando que o grupo D e o grupo E tiveram notas maiores em matematica

#agrupando o nivel de escolaridade e plotando com a media das notas em escrita
